```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load dataset
df = pd.read_csv("mobiles1.csv")

# Data Cleaning
# Fill missing values
df = df.copy()
df.loc[:, "processor"] = df["processor"].fillna("Unknown Processor")
df.loc[:, "rating"] = df["rating"].fillna(df["rating"].median())
df.loc[:, "reviews"] = df["reviews"].fillna("0 Reviews")
df.loc[:, "warranty"] = df["warranty"].fillna("No Warranty Info")

# Convert reviews to numeric
df.loc[:, "reviews"] = df["reviews"].str.replace(r"[^\d]", "", regex=True).astype(int)

```

```python
# Set plot style
sns.set_style("darkgrid")

# Price Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df["price"], bins=30, kde=True, color="blue")
plt.title("Price Distribution of Mobiles")
plt.xlabel("Price (INR)")
plt.ylabel("Count")
plt.show()

# Rating Distribution
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["rating"], color="green")
plt.title("Boxplot of Mobile Ratings")
plt.xlabel("Rating")
plt.show()

# Reviews vs. Price
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["price"], y=df["reviews"], alpha=0.5, color="red")
plt.title("Number of Reviews vs. Price")
plt.xlabel("Price (INR)")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
# Insights
print("Insights:")
print("1. Price Distribution: Most mobile prices are concentrated in the lower range, with fewer high-priced models.")
print("2. Rating Distribution: Most ratings are between 4.0 and 4.6, indicating generally positive reviews.")
print("3. Reviews vs. Price: No strong correlation; some budget phones have high reviews while some expensive ones have fewer reviews.")
```
