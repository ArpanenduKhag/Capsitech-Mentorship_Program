```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the dataset
file_path = "investment_data.csv"  # Ensure correct file path if running in Colab
df = pd.read_csv(file_path)

# Handle missing values
df.dropna(inplace=True)
```

```python

# Summary statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python

# Distribution of investments across categories
plt.figure(figsize=(12, 6))
df.drop(columns=['category_list']).sum().sort_values().plot(kind='barh', color='skyblue')
plt.xlabel('Total Investments')
plt.ylabel('Categories')
plt.title('Total Investments Across Categories')
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.drop(columns=['category_list']).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

# Boxplot to analyze investment distribution
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.drop(columns=['category_list']))
plt.xticks(rotation=90)
plt.title('Investment Distribution Across Categories')
plt.show()
```

```python

# Insights:
print("\nInsights:")
print("1. The most invested category appears to be [identify from the bar chart].")
print("2. Some categories have highly skewed distributions, as seen in the boxplot.")
print("3. The correlation heatmap shows strong/weak relationships between certain categories.")
```
